```python
import pandas as pd
import sqlite3
```

```python
with sqlite3.connect('data/games.db') as con:
    games = pd.read_sql(sql='SELECT * FROM games', con=con)

games
```

```python
games = games.dropna()
games = games[games['event'] == 'NBA']
games = games[games['season'] < 2021]
games = games[games['home_odds'] != '-']
games = games[games.home_odds != games.away_odds]
games['home_team'] = games['home_team'].str.strip()
games['away_team'] = games['away_team'].str.strip()
games['date'] = pd.to_datetime(games['date']) - pd.Timedelta('02:00:00')

games
```

```python
games = games[['home_score', 'away_score', 'home_odds', 'away_odds']]
games
```

```python
games['score'] = games['home_score'] - games['away_score']
games['home_odds'] = pd.to_numeric(games['home_odds'])
games['odds'] = games['home_odds'] - games['away_odds']
```

```python
games
```

```python
games['result'] = (games['score'] > 0).astype(int)
```

```python
games.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
```

```python
x = games.odds
y = games.result

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
sgd = SGDClassifier()
```

```python
sgd.fit(pd.DataFrame(x_train), y_train)
```

```python
score = sgd.score(pd.DataFrame(x_test), pd.DataFrame(y_test))
score
```
