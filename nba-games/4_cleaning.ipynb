```python
import pandas as pd
import sqlite3
```

```python
with sqlite3.connect('data/games.db') as con:
    games = pd.read_sql(sql='SELECT * FROM games', con=con)

games
```

```python
games = games.dropna()
games
```

```python
games = games[games['event'] == 'NBA']
games
```

```python
games['date'] = pd.to_datetime(games['date'])
games
```

```python
games['date'] = games['date'] - pd.Timedelta('02:00:00')
```
